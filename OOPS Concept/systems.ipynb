Question 1 :- Bank Account System

```python
class Account:
    def __init__(self, name, acc_no, balance=0):
        self.__name = name
        self.__acc_no = acc_no
        if balance < 0:
            balance = 0
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Welcome {self.__name} Your bank account number {self.__acc_no} successfully credited {amount}. Now total amount is {self.__balance}")
        else:
            print("Amount should be positive!")

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient Balance!")
        else:
            self.__balance -= amount
            print(f"Welcome {self.__name} {amount} debited successfully from your account number {self.__acc_no}. Now the current balance in your account is {self.__balance}")

    def check_balance(self):
        print(f"Welcome {self.__name} Your account number {self.__acc_no} has a balance of {self.__balance}")

somil_acc = Account("Somil", "056010052060", 100)
somil_acc.check_balance()
somil_acc.deposit(500)
somil_acc.withdraw(200)
```

Question 2 :- Employee Management System

```python
class Employee:
    def __init__(self, name, salary, company_name):
        self._name = name
        self._salary = salary if salary >= 0 else 0
        self._company_name = company_name

    def display_info(self):
        print(f"I am an Employee in {self._company_name}")

class Manager(Employee):
    def __init__(self, name, salary, company_name, team_size):
        super().__init__(name, salary, company_name)
        self._team_size = team_size

    def display_info(self):
        print(f"I am a Manager at {self._company_name}. I have a team of {self._team_size}")

class Developer(Employee):
    def __init__(self, name, salary, company_name, programming_language):
        super().__init__(name, salary, company_name)
        self._programming_language = programming_language

    def display_info(self):
        print(f"I am a Developer at {self._company_name} and I am proficient in {self._programming_language}")

somil = Developer("Somil", 200000, "Google", "Java")
somil.display_info()
ayan = Manager("Ayan", 150000, "Microsoft", 7)
ayan.display_info()
```

Question 3 :- Library System

```python
class Library:
    def __init__(self, name):
        self.__name = name
        self.__books = {}  # {book_id: {'title': '', 'author': '', 'available': True}}
        self.__issued_books = {}  # {member_id: [book_ids]}
        self.__book_counter = 0

    def add_book(self, title, author, quantity=1):
        for _ in range(quantity):
            self.__book_counter += 1
            self.__books[self.__book_counter] = {
                'title': title,
                'author': author,
                'available': True
            }
        print(f"Added {quantity} copy/copies of '{title}' to {self.__name}")

    def issue_book(self, member_id, book_id):
        if book_id not in self.__books:
            print("Book not found in library!")
            return False
        
        if not self.__books[book_id]['available']:
            print(f"'{self.__books[book_id]['title']}' is not available right now!")
            return False
        
        self.__books[book_id]['available'] = False
        if member_id not in self.__issued_books:
            self.__issued_books[member_id] = []
        self.__issued_books[member_id].append(book_id)
        print(f"Book '{self.__books[book_id]['title']}' issued to member {member_id}")
        return True

    def return_book(self, member_id, book_id):
        if member_id not in self.__issued_books or book_id not in self.__issued_books[member_id]:
            print("This book was not issued to this member!")
            return False
        
        self.__books[book_id]['available'] = True
        self.__issued_books[member_id].remove(book_id)
        print(f"Book '{self.__books[book_id]['title']}' returned by member {member_id}")
        return True

    def show_available_books(self):
        available = [book for book_id, book in self.__books.items() if book['available']]
        if not available:
            print("No books available in the library!")
            return
        
        print(f"\n--- Available Books in {self.__name} ---")
        for book_id, book in self.__books.items():
            if book['available']:
                print(f"ID: {book_id} | Title: {book['title']} | Author: {book['author']}")

# Create library instance
lib = Library("Central Library")
lib.add_book("Python Programming", "Guido van Rossum", 3)
lib.add_book("Data Science", "John Doe", 2)
lib.add_book("Web Development", "Jane Smith", 1)

lib.show_available_books()
lib.issue_book(101, 1)
lib.show_available_books()
lib.return_book(101, 1)
lib.show_available_books()
```
