Question 1 :- create a line chart with days and sales

```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5]
sales = [100,150,120,170,200]

plt.plot(days, sales)
plt.title("Sales Graph")
plt.xlabel('Days')
plt.ylabel('Sales')

plt.show()
```
