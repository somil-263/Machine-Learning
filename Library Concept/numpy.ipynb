Question 1 :- Basic Statistics

```python
import numpy as np
```

```python
arr = np.array([10, 20, 30, 40, 50])

print("Mean of array is: ", np.mean(arr))
print("Median of array is: ", np.median(arr))
print("Standard Deviation of array is: ", np.std(arr))
print("Variance of array is: ", np.var(arr))
```

Question 3 :- Boolean Indexing

```python
arr = np.array([5, 10, 15, 20, 25, 30])

arr2 = np.array(arr > 15)
print(arr2)
```
