Question 1 :- Dataframe Filtering

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "name":["Rahul","Aman","Priya","Ankit"],
    "age":[21,19,22,20],
    "salary":[50000,60000,55000,70000]
})
```

```python
print(df[df['age'] > 20])
print(df[df['salary'] > 55000])
print(sum(df['salary'])/4)
```

Question 2 :- Use of Group By and find the average salary per department

```python
df = pd.DataFrame({
    'department' : ['IT','IT','HR','HR','Sales'],
    'salary' : [50000,60000,45000,55000,70000]
})
```

```python
result = df.groupby('department')['salary'].mean().to_frame()
for department, salary in result['salary'].items():
    print(f"{department} -> {salary}")
```

Question 3 :- Missing Values

```python
df = pd.DataFrame({
    "name":["A","B","C","D"],
    "marks":[90,None,80,None]
})
```

```python
df['marks'].fillna(df['marks'].sum()/4, inplace=True)
```
