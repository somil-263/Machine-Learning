Question 2 :- print the average number of user input and stop if exit print and if try to enter any invalid input then throw the exception program should not be crashed

```python

num = 0
count = 0
while True:
    try:
        val = input("Enter the number")
        if(val == "exit1"): break
        num += int(val)
        count += 1

    except Exception as ex:
        print(ex)

print(num / count)
```

Question 4 :- print the most frequency word in the data.txt

```python
with open('data.txt', 'r') as file:
    try:
        data = file.read()
        lst = data.split()
        dict = {}
        for word in lst:
            dict[word] = dict.get(word, 0) + 1
        
        max_freq = max(dict.values()) if dict else 0
        most_frequent = [k for k, v in dict.items() if v == max_freq]
        for word in most_frequent:
            print(word, "-> ", max_freq)
    except Exception as ex:
        print(ex)
```
