Question 1 :- Student Grade Calculator

```python
def grade_calculator(marks):
    if marks > 90:
        return 'A'
    elif marks > 80:
        return 'B'
    elif marks > 70:
        return 'C'
    else:
        return 'F'

student_marks = [94, 32, 57, 93, 75, 65, 82, 79, 90, 100]

student_grades = list(map(grade_calculator, student_marks))

print(student_grades)
```

Question 2 :- Sort the details in descending order

```python
students = [
    ("Rahul", 85),
    ("Aman", 92),
    ("Priya", 78),
    ("Ankit", 95)
]

students.sort(key=lambda x: x[1], reverse=True)
print(students)
```

Question 3 :- Filter the even number

```python
nums = [12, 3, 17, 25, 40, 9]

even_nums = list(filter(lambda x : x%2 == 0, nums))

print(even_nums)
```
