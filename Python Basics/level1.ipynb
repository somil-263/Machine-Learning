Question 1 :- Take user input and check wheather it is even or odd

```python
num = int(input("Enter The Number"));
if num < 0:
    print("Number is negative")
elif num % 2 == 0:
    print("Number is even")
else:
    print("Number is Odd")
```

Question 4 :- Take input as year and check whether it is leap or not

```python
year = int(input("Enter Year"))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("This is Leap Year")
        else:
            print("This is not leap year")
    else:
        print("This is leap year")

else :
    print("This is not a leap year")
```
