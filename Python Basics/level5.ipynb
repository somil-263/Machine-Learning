Question 1 :- print the frequency of characters in the string

```python
str = "programming"

dict = {}

for ch in str:
    if ch in dict:
        dict[ch] = dict.get(ch) + 1
    else :
        dict[ch] = 1

print(dict)
```

Question 2 :- Unique Words

```python
str = "I love python and I love machine learning"

lst = str.split(' ')
lst2 = []

for word in lst:
    if word in lst2:
        continue
    else :
        lst2.append(word)

print(lst2)
```

Question 3 :- Dictionary Inversion

```python
dict = {
    'a' : 1,
    'b' : 2,
    'c' : 3
}

dict2 = {}

for key, value in dict.items():
    dict2[value] = key

print(dict2)
```

Question 4 :- Filter the name starts with a

```python
names = ["Rahul", "Ankit", "Aman", "Priya", "Arjun", "ayan"]

res = [name for name in names if name.startswith('A') or name.startswith('a')]

print(res)
```
