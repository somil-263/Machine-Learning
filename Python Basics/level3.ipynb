Question 2 :- Check whether string is palindrome or not

```python
text = input("Enter the string")

if(text == text[::-1]):
    print("Palindrome")
else:
    print("Not Palindrome")
```

Question 3 :- count the vowel in text

```python
text = input("Enter the text")
vowelCount = 0
vowels = "aeiouAEIOU"

for character in text:
    if(character in vowels):
        vowelCount += 1

print(vowelCount)
```
