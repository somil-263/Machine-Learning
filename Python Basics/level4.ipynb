Question 1 :- Remove the Duplicates from the list

```python
nums = [1, 2, 2, 3, 4, 4, 5, 1]

st = set(nums)

print(st)
```

Question 2 :- Print the common elements from 1 list

```python
a = [1, 2, 3, 4, 5]
b = [4, 5, 6, 7]

st = set(a).intersection(set(b))

print(st)
```

Question 3 :- Make the list from 1 to 100 all even numbers

```python
lst = [i for i in range(1, 101) if i % 2 == 0]

print(lst)
```

Question 4 :- Make the Square numbers dictionary

```python
dict = {i: i*i for i in range(1, 101) if i % 2 == 0}

print(dict)
```
