Question 1 :- Expense Tracker

```python
expenses = {}
while True:
    inp = input("Enter the Expense: ")
    if inp.strip() == "":
        break

    parts = inp.rsplit(maxsplit=1)
    if len(parts) != 2:
        print("Please enter item and amount, e.g. 'chicken korma 300'")
        continue

    name, amount_str = parts
    try:
        amount = int(amount_str)
    except ValueError:
        print("Amount must be a number")
        continue

    expenses[name] = expenses.get(name, 0) + amount

print(expenses)
```

Question 2 :- Contact Book

```python
phone_book = {}

def add_contact(name, number):
    name = name.lower()
    phone_book[name] = number

def search_contact(name):
    name = name.lower()
    if name in phone_book:
        print(phone_book[name])
    else:
        print(f"{name.capitalize()} is not exist in the phone book")

def delete_contact(name):
    name = name.lower()
    if name in phone_book:
        del phone_book[name]
    else:
        print(f"{name.capitalize()} is not exist in the phone book")

def show_contact():
    for name, number in phone_book.items():
        print(f"{name.capitalize()}: {number}")

show_contact()
add_contact('Somil', 9627434602)
add_contact('Ayan', 9149127725)
add_contact('Kali', 9412126345)
add_contact('King', 9178678612)
show_contact()
search_contact("Somil")
search_contact("somil")
search_contact("sOmil")
search_contact("somil chauhan")
delete_contact("King")
show_contact()
```

Question 3 :- Data Cleaning

```python
data = [10, None, 20, '', 30, None, 40, 'somil']

data = [x for x in data if isinstance(x, int)]
print(data)
```

Question 4 :- Frequency Counter

```python
data = [1,2,3,2,1,4,2,5,1]
freq = {}

for num in data:
    freq[num] = freq.get(num, 0) + 1

print(freq)
```
