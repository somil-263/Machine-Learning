Question 1 :- print sum from nth natural number

```python
n = int(input("Enter nth value"))

sum = 0

for i in range(n+1):
    sum += i

print(sum)
```

Question 2 :- print factorial of given number

```python
num = int(input("Enter the number"))
fact = 1;

for i in range(1, num+1):
    fact *= i

print(fact)
```

Question 3 :- Print the Multiplication Table of given number

```python
table = int(input("Enter the number"))

for i in range(1, 11):
    print(f"{table} x {i} = {table * i}")
```

Question 4 :- Check the number is prime or not

```python
num = int(input("Enter the number"))
flag = True

for i in range(2, num):
    if(num % i == 0):
        flag = False

if flag:
    print("Number is Prime")
else:
    print(f"{num} is not prime")
```
