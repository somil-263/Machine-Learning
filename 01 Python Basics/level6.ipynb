Question 1 :- Group words by length

```python
words = ["cat", "dog", "elephant", "bat", "tiger"]

dict = {}

for word in words:
    n = len(word)
    if n in dict:
        dict[n].append(word)
    else:
        dict[n] = [word]

print(dict)
```

Question 2 :- Find missing value

```python
lst1 = [1,2,3,4,6,7,8]

st = lst1[0]
end = lst1[-1]

lst2 = [i for i in range(st, end+1)]

res = set(lst2).difference(set(lst1))

print(res)
```

Question 3 :- Find:

Highest math score student
Average science marks

```python
students = {
    "A": {"math": 90, "science": 80},
    "B": {"math": 70, "science": 95},
    "C": {"math": 85, "science": 88}
}

mathMax = 0
scienceTotal = 0

for i in students.keys():
    mathMax = max(mathMax, students[i].get("math"))
    scienceTotal += students[i].get("science")

print(f"Maximum Math marks are {mathMax}")
print(f"The average marks of science are {scienceTotal / len(students)}")
```

Question 4 :- word frequency

```python
sentence = "The quick brown fox jumps over the lazy dog the fox"
lst = sentence.split(' ')
res = {}

for word in lst:
    if word.lower() in res:
        res[word.lower()] = res.get(word.lower()) + 1
    else:
        res[word.lower()] = 1

print(res)
```

Question 5 :- numbers > 20 and squares of all numbers and even numbers only

```python
data = [12, 15, 18, 20, 22, 25, 30]

res = [num ** 2 for num in data if num > 20 and num % 2 == 0]

print(res)
```
